```python
df.groupby('category')[['sales', 'profit']].sum().sort_values('sales', ascending = False)
```

```python
region_summary = df.groupby('region')[['sales','profit']].sum()
region_summary['profit_status'] = region_summary['profit'].apply(lambda x: 'Loss' if x < 0 else 'Profit')
region_summary.sort_values('profit')
```

```python
def discount_level(discount):
    if discount == 0:
        return 'No Discount'
    elif discount <= 0.20:
        return 'Low Discount'
    else:
        return 'High Discount'

df['discount_level'] = df['discount'].apply(discount_level)
df.groupby('discount_level').agg(total_orders=('discount','count'), avg_profit=('profit','mean'), total_profit=('profit','sum')).sort_values('avg_profit', ascending=False)
```

```python
subcat_profit = df.groupby('sub_category')['profit'].sum()
subcat_profit[subcat_profit < 0].sort_values()
```

```python
def customer_value(profit):
    if profit >= 90:
        return 'High Value'
    elif profit >= 50:
        return 'Average Value'
    else:
        return 'Low Value'

segment_summary = df.groupby('segment').agg(
    total_sales=('sales','sum'),
    avg_order_value=('sales','mean'),
    avg_profit_per_order=('profit','mean')
)
segment_summary['customer_value'] = segment_summary['avg_profit_per_order'].apply(customer_value)
segment_summary.sort_values('avg_profit_per_order', ascending=False)
```
